# Compare GDP for the top 10 Populous Countries in 2007

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_2007 = data[data.year ==2007]
data_2007
```

```python
#get the ten most popuplous countries using the sort function 
top10 = data_2007.sort_values('population', ascending = False).head(10)
top10
```

```python
plt.subplot(2,1,1)
x = range (10)
#ADD POPULATION GRAPH AS A SUBPLOT
plt.bar(x,top10.population/10**6)
#remove the lables
plt.xticks([],[])
plt.title('GDP FOR 10 TEN POPUPLOUS COUNTRIES')
plt.legend(['Population'])


plt.subplot (2,1,2)

plt.bar (x,top10.gdpPerCapita * top10.population/10**9, color ='red')
plt.xticks(x,top10.country,rotation = 'vertical')
plt.xlabel('country')
plt.ylabel('GDP')
plt.legend(['GDP'])



plt.show()
```
